```python
import pandas as pd
```

```python
def find_cat(desc, cat_dict):
    cat_list = []
    for cat in cat_dict:
        for w in cat_dict[cat]:
            if w in desc:
                cat_list.append(cat)
    return cat_list

cat_d = {
    "Arts and Entertainment": ["artwork","pub","memorial"
    ,"amusementpark","casino","cinema","theatre","museum","bowling"],

    "Business and Professional Services": ["bank", "hotels","banking and finance",
    "financial service"],

    "Retail": ["restaurant", "food","shop","bakery",
    "hair","convenience","bar","any","laundry",
    "pizzeria","store","supermarket","clothes","burger",
    "cafe","butcher","retail","florist","beverages",
    "breakfast spot","snack","creperie","gastropub","greengrocer",
    "steakhouse","lounge"],

    "Community and Government": ["post","atm","police",
    "newsagent","community","wash","doityourself"],

    "Education": ["kindergarten","college","school"],

    "Landmarks and Outdoors": ["plaza", "playground","park","bridge",
    "pitch","fountain","sports","pedestrian plaza","swimming pool",
    "stadium","attraction","roof deck","landmarks and outdoors","diner",
    "scenic","garden","fish market","dealership","food court","brasserie",
    "viewpoint","basketball"],

    "Health and Medicine": ["optician","pharmacy","dentist","doctors",
    "drugstore","veterinary","health"],

    "Travel and Transportation":["tourist_info","parking","rental","bicycle","bicyclerental"]

        }

df = pd.read_csv('MergePoiDel.csv')
df["Venue Category"] = df[["fclass"]].apply(lambda row: find_cat(row["fclass"], cat_d), axis=1) 
df.to_csv('MergePoiDelSorted.csv')
```
