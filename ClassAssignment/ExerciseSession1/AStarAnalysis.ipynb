## Background - Artificial Intelligence

### Problems Solving

* Using an appropriate **state space** to represent the problem.

* To test whether the **target state** appears in the state space.
<br/>

The purpose of intelligent system is to find the **action sequence** that can achieve the **desired goal**, and make it pay the **least cost** and have the **best performance**.
<br/>


### Problem representation

Four steps to represent problems

* Initial state set: defines the initial environment;

* Operator set: the action of changing a problem from one state to another;

* Target detection function: used to determine whether a state is a target;

* Path cost function: a function that assigns a certain cost to each path.



#### State Space Problem

Two dimension State and operation

* State  
Status is a **data structure** that represents the **status of each step** in the process of problem solving. It can be expressed in the following form   
<br/> S_k \ = \ \{S_{k0},S_{k1},...,S_{kn}\} 
<br/> 

* Operation   
Operation can also be called operator, which is a means to **change the problem from one state to another**.
<br/>

* State Space    
The state space is a set **composed of all States of a problem and the relationship between these states**. A triple can be used:  
<br/>（S,F,G)<br/>   
To show. Where S is the set of all initial states of the problem, F is the set of operations, and G is the set of target states.

### Search Strategy

Features of blind search:

* The search is carried out according to the **specified route without using heuristic information** related to the problem;

* Heuristic search should use **heuristic information related to the problem**, and use these heuristic information to **guide the search process**.


* The deficiency of blind search: **low efficiency, consuming too much computing space and time**.

* Analyze the width first, depth first search, or cost search algorithms introduced earlier. The **main difference** is the **order of nodes** to be expanded in the open table. People try to **find a way to arrange** the order of nodes to be expanded, that is, select the most promising nodes to expand, then the search **efficiency will be greatly improved**.

* Heuristic information: when searching, you can use some **information about specific problem areas**, which is called **heuristic information**.

* The search method using **heuristic information** is called heuristic search method.

## Principle of A*

###  The Search Area
* We have someone who wants to get from point A(green) to point B(red), but a wall(bule) separates the two points.  
* Divide our search area into a square grid.(This could be rectangles, hexagons, triangles, or any shape.)

![](https://uploads.gamedev.net/monthly_2018_05/aStarT1.jpg.d75d82a7b81262ca3d882353c92d0f51.jpg)

### Starting the Search
* Begin at the starting point A and add it to an open list.
* Look at all the reachable or walkable squares adjacent to the starting point:
  * Squares with walls: ignore them.
  * Not searched square: add them to the open list. For each of these squares, save point A as its "parent square". 
* Drop the starting square A from open list, and add it to a closed list.  

![](https://uploads.gamedev.net/monthly_2018_05/aStarT2.jpg.461d6c92a13dad908f94279636ba18d8.jpg)

### Path Sorting
* Figuring out the path: F = G + H
  * G = the movement cost to move from the starting point A to a given square on the grid, following the path generated to get there.
    * Assign a cost of 10 to each horizontal or vertical square moved, and a cost of 14 for a diagonal move.
  * H = the estimated movement cost to move from that given square on the grid to the final destination, point B.
    *  Manhattan method: calculate the total cost to move horizontally and vertically to reach the target square from the current square, ignoring diagonal movement, and ignoring any obstacles.
    
![](https://imgconvert.csdnimg.cn/aHR0cDovL3d3dy5jcHBibG9nLmNvbS9pbWFnZXMvY3BwYmxvZ19jb20vY2hyaXN0YW54dy9pbWFnZTAwMy5qcGc?x-oss-process=image/format,png)

### Continuing the Search
* Choose the lowest F score square from all those that are on the open list.
* Drop the chosen square from open list, and add it to a closed list.
* Look at all the reachable or walkable squares adjacent to the chosen square:
  * Squares in closed list: ignore them.
  * Squares with walls: ignore them.
  * Not searched square: add them to the open list. For each of these squares, save the chosen square as its "parent square". 
  * Square already in the open list: check whether this path is better, that is, whether the current square has a smaller G score:
    * If not: do nothing.
    * If so: switche its "parent square" to chosen square, recalculate G and F scores.
    
![](https://imgconvert.csdnimg.cn/aHR0cDovL3d3dy5jcHBibG9nLmNvbS9pbWFnZXMvY3BwYmxvZ19jb20vY2hyaXN0YW54dy9pbWFnZTAwNC5qcGc?x-oss-process=image/format,png)
<br/>

* Repeat this process.

![](https://uploads.gamedev.net/monthly_2018_05/aStarT5.jpg.b2beeb270ef7f5f9603f26314775b3e1.jpg)
<br/>


* Until target square is added to the closed list.



### Determine the path
* Start from the red target square, and work backwards moving from one square to its parent, following the arrows("parent square").
