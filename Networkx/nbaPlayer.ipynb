```python
class Player:
    def __init__(self, name, state, collage):
        self.name = name
        self.state = state
        self.collage = collage
```

```python
nbaPlayers = []
with open("Players.csv") as PlayerData:
    reader = csv.DictReader(PlayerData)
    for row in reader:
        name = row['name']
        state = row['state']
        collage = row['collage']
        nbaPlayers.append(Player(name, state, collage))
```

```python
print(nbaPlayers[20].name)
print(nbaPlayers[20].state)
```

```python
PlayerGraph = nx.Graph()
```

```python
playerInfo = nbaPlayers[0:100]
```

```python
for f in playerInfo:
    PlayerGraph.add_node(f,name = (f.name[:15]))
```

```python
for f1 in playerInfo:
    for f2 in playerInfo:
        if not f1 == f2:
            if f1.state == f2.state:
                PlayerGraph.add_edge(f1,f2)
```

```python
for f3 in playerInfo:
    for f4 in playerInfo:
        if not f3 == f4:
            if f3.collage == f4.collage:
                PlayerGraph.add_edge(f3,f4)
```

```python
labels = nx.get_node_attributes(PlayerGraph, 'name')
print(labels)
```

```python
options = {'node_color':'blue', 'node_size':100, 'edge_color':'coral','width':3, 'labels':labels, 'with_labels': True}
plt.figure(1, figsize = (60, 60))
nx.draw(PlayerGraph, **options)
```

### This graph showed the coonections when players own the same birth-state or collage.
